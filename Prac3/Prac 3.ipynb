```python
import nltk
import string
import re
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import os
```

```python
ans = []
path = '/content/doc1.txt'
myFile = open(path,"r")
ans.append(myFile.read())

path = '/content/doc2.txt'
myFile = open(path,"r")
ans.append(myFile.read())

path = '/content/doc3.txt'
myFile = open(path,"r")
ans.append(myFile.read())

path = '/content/doc4.txt'
myFile = open(path,"r")
ans.append(myFile.read())

path = '/content/doc5.txt'
myFile = open(path,"r")
ans.append(myFile.read())

path = '/content/doc6.txt'
myFile = open(path,"r")
ans.append(myFile.read())

total_text = "".join(i for i in ans)
```

```python
total_text.lower()
```

```python
total_text = "".join([char for char in total_text if char not in string.punctuation])
```

```python
words = word_tokenize(total_text)
```

```python
stop_words = stopwords.words('english')
filtered_words = [word for word in words if word not in stop_words]
```

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
```

```python
frequency_distribution = nltk.FreqDist(word for word in (total_text and stemmed_words))
```

```python
frequency_distribution
```

```python
freq_dict = {}
for word, frequency in frequency_distribution.most_common(1000):
    freq_dict[word] = frequency
```

```python
import matplotlib.pyplot as plt

plt.xticks(rotation=90)
plt.bar(list(freq_dict.keys())[:30], list(freq_dict.values())[:30])
plt.show()
```
